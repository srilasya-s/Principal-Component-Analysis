```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
y = iris.target

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title("Iris Dataset PCA with 2 Components")
plt.xlabel("PrincipalComp 1")
plt.ylabel("PrincipalComp 2")
plt.colorbar().set_ticks(np.unique(y))
# Answer for question a
plt.show()
```

```python
#Answer for question b
var_cap = pca.explained_variance_ratio_.sum() * 100
print(f"Total percentage of variance captured by the first 2 PCA components: {var_cap:.2f}%")
```

```python
# Answer for question c
print("Strength of first 2 PCA components:")
print(pca.explained_variance_ratio_)
```

```python
#Answer for question d
print("Magnitude of first 2 PCA components:")
print(pca.explained_variance_)
```
